```python
#Version 1

times_found = 0 #Variable de conteo de las veces que aparece el target en secuence
comparing_index = 0 #Indice de conteo dentro de la comparacion
target = 'AAAATTTATTGTCC' #Target grupo 2
location = [] # Arreglo de posiciones de interes
#Todo este bloque es de O(1) dado que asignaciones son de este orden

for i in range(len(secuence)): #O(len(secuence)) = O(n)
  if target[comparing_index] == secuence[i]: #Comparacion indice de la seuencia con indice del target
    comparing_index += 1
  else:
    comparing_index = 0
  #O(1)
  if comparing_index == len(target): #Condicion para incluir en location
    times_found += 1
    comparing_index = 0
    location.append(i - len(target)+1)
  #O(1)
#El orden final de este bucle es de O(n) dado que O(n)*O(1) = O(n)

print("The strig was found exactly the following times in the covid-19 genom: ", end = ' ')
print(times_found)
print("Location of the string starting to count from 0", end = ' ')
print(location)
for i in range(len(location)):
  location[i] += len(target)-1
print("End location of the string", location)
#O(1) de los print
#O(len(location)) para el bucle final

#El orden coste total del algoritmo es de O(n)
```

```python
#Version 2

times_found = 0 #Igual que en algoritmo anterior
target = 'AAAATTTATTGTCC' #Target grupo 2
location = [] #Igual q en algoritmo anterior
#Todo este bloque es de O(1) dado que asignaciones son de este orden

for i in range(len(secuence)): #O(len(sequence)) = O(n)
  if secuence[i:i+len(target)] == target: #O(1)
    times_found += 1
    location.append(i)
#El orden final del bucle es de O(n)*O(1) = O(n)

print("The strig was found exactly the following times in the covid-19 genom: ", end = ' ')
print(times_found)
print("Locations of the string", end = ' ')
print(location)
for i in range(len(location)):
  location[i] += len(target)-1
print("End locations of the string", location)
#O(1) de los print
#O(len(location)) para el bucle final

#El orden coste total del algoritmo es de O(n)
```
